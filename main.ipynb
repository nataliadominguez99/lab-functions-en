## 2. Now write a function that returns the largest element on a list

```python
numbers = [1,5,7,8,6,2,25,30]
```

```python
def greatest(numbers):
    #start by 0
    max_number = numbers [0]

    for number in numbers:
        if number > max_number:
            max_number = number
    return max_number
```

## 6. Write a function that returns the factorial of a number.

```python
def factorial(n):
    if n < 0:
        return 1  #undefined for negative numbers
    result = 1
    i = 1
    while i <= n:
        result = result * i
        i = i + 1
    return result
```

```python
#factorial formula
#n! = n * ( n - 1 ) *...*1

# This code defines a function called "factorial" which takes an input "n". The function uses a for loop to iterate through the range of numbers 
# from 1 to n+1. For each number in that range, it multiplies the current value of x by the number in the range. At the end of the loop, 
# the function returns the final value of x, which will be the factorial of the input number "n".

# The Factorial of a positive integer n is the product of all positive integers less than or equal to n. 
# For example, the factorial of 6 (written "6!") is 6 * 5 * 4 * 3 * 2 * 1 = 720.

# So this function takes an input of any positive integer, and returns the factorial of that number.
```

```python
# This will test your function 
test_factorial(factorial)
```

## 7. Write a function that takes a list and returns a list of the unique values.

`NOTE: You cannot use set. ðŸ¤”`

```python
def unique(lst_un):
#create empty list
    unique_list = []

    #search for the value in the principal list
    for value in lst_un:
        #if the value is not in the unique list
        if value not in unique_list:
            #add value to the empty list
            unique_list.append(value)
        
    return unique_list
    
```

## 8. Write a function that returns the mode of a list, i.e.: the element that appears the most times.
`NOTE: You should not use count... ðŸ§`

```python
def mode_counter(arr):
    # Initialize mode as an empty list to accommodate multiple modes
    mode = []
    counts = {}
    
    # Count occurrences of each element
    for i in arr:
        if i in counts:
            counts[i] += 1
        else:
            counts[i] = 1

    # Find the maximum count
    first_value_found = False
    for count_val in counts.values():
        if not first_value_found:
            max_count = count_val
            first_value_found = True
        elif count_val > max_count:
            max_count = count_val

    # Collect all items with the maximum count
    for item, count in counts.items():
        if count == max_count:
            mode.append(item)
```

## 9. Write a function that calculates the standard deviation of a list.
`NOTE: Do not use any libraries or already built functions. ðŸ˜‰`

```python
def st_dev(list_sd):
    # Step 1: Calculate the mean
    total = 0                 # Initialize total sum to 0
    count = 0                 # Initialize count of numbers to 0
    for num in list_sd:
        total += num          # Add current number to total sum
        count += 1            # Increment count for each number encountered
    mean = total / count      # Calculate mean (average) of the list

    # Step 2: Calculate the variance
    sum_squared_diff = 0      # Initialize sum of squared differences to 0
    for num in list_sd:
        deviation = num - mean      # Calculate difference from mean
        squared_deviation = deviation * deviation  # Square the difference
        sum_squared_diff += squared_deviation      # Add squared difference to total sum
    variance = sum_squared_diff / count  # Calculate variance as average of squared differences
    
    # Step 3: Compute the standard deviation
    # Without using libraries, we'll manually compute the square root
    # Approximating square root can be simplified with ** 0.5
    standard_deviation = variance ** 0.5

    return standard_deviation
```

## 10. Write a function to check if a string is a pangram, i.e.: if it contains all the letters of the alphabet at least once. Mind that the strings may contain characters that are not letters.

```python
def is_pangram(s):

    alphabet_set = set()  # Initialize an empty set to keep track of letters
    
    #Iterate over each character in the string
    for char in s:
        # Convert character to lowercase to ensure case does not matter
        char = char.lower()
        
        # Check if the character is a letter
        if char.isalpha():
            alphabet_set.add(char)  # Add the letter to the set
    
    # Check if the set size is 26 (i.e., all letters present)
    return len(alphabet_set) == 26
```

```python
# This will test your function 
test_pangram(is_pangram)
```

## 11. Write a function that receives a string of comma separated words and returns a string of comma separated words sorted alphabetically.

`NOTE: You may use sorted but not split and definitely no join! ðŸ¤ª`

```python
def sort_alpha(string):

    words = []  # List to hold the extracted words
    current_word = ""  # Variable to build each word
    
    # Extract words from the string
    for char in string:
        if char == ',':  # A comma indicates the end of a word
            words.append(current_word)  # Add the completed word to the list
            current_word = ""  # Reset current word for the next
        else:
            current_word += char  # Add current character to the word
    
    # To handle the last word after the final comma
    words.append(current_word) 
    
    # Sort words alphabetically
    sorted_words = sorted(words)  # Use the sorted function
    
    # Create a sorted string without using join
    sorted_string = ""
    for i, word in enumerate(sorted_words):
        if i > 0:  # Add a comma before from the second word onward
            sorted_string += ','
        sorted_string += word  # Add the current word

    return sorted_string
```

```python
# This will test your function 
test_alpha(sort_alpha)
```

## 12. Write a function to check if a given password is strong (at least 8 characters, at least one lower case, at least one upper case, at least one number and at least one special character). It should output True if strong and False if not.

```python
def check_pass(password):

    #Check if the length is at least 8 characters
    if len(password) < 8:
        return False  # If password is shorter than 8 characters, it's not strong

    has_lower = False  # Flag to check for lowercase letter
    has_upper = False  # Flag to check for uppercase letter
    has_digit = False  # Flag to check for a digit
    has_special = False  # Flag to check for a special character
    
    # Define special characters in the password
    special_characters = "!@#$%^&*()-_=+[{]}\\|;:'\",<.>/?"
    
    #Check each character in the password
    for char in password:
        if char.islower():
            has_lower = True  # Found a lowercase letter
        elif char.isupper():
            has_upper = True  # Found an uppercase letter
        elif char.isdigit():
            has_digit = True  # Found a digit
        elif char in special_characters:
            has_special = True  # Found a special character
        
        # If all conditions are met, return True immediately
        if has_lower and has_upper and has_digit and has_special:
            return True
    
    # If any of the character type requirements are not met, return False
    return False
```

```python
# This will test your function 
test_pass(check_pass)
```
